```python
#Функции

"""
Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, 
которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:
код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, 
дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу 
(вводом команды “q”).
"""

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}


def main():
    while True:
        user_input = input('Введите команду (p,s,l,as,ds,ad,d,m,q): ').lower()
        if user_input == 'p':
            print(print_document_owner(documents))
        elif user_input == 's':
            print(print_shelf(directories))
        elif user_input == 'l':
            print(print_docs_info(documents, directories))
        elif user_input == 'as':
            print(add_new_shelf(directories))
        elif user_input == 'ds':
            print(delete_empty_shelf(directories))
        elif user_input == 'ad':
            print(add_new_document(documents, directories))
        elif user_input == 'd':
            print(delete_document(documents, directories))
        elif user_input == 'm':
            print(move_doc_to_shelf(documents, directories))
        elif user_input == 'q':
            print('До свидания!')
            break

#Задание 1
#Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру

def print_document_owner(docs):
    number_document = input('Введите номер документа: ')
    for i in range(0,len(documents)):
        if documents[i]['number'] == number_document:
            print ('Результат:\nВладелец документа: ', documents[i]['name'])
            return print_document_owner(docs)
    
    if documents[i]['number'] != number_document: 
        print ('Результат:\nДокумент не найден в базе!')  
        return main()  


#Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
def print_shelf(dirs):
    number_document = input('Введите номер документа: ')
    for shelf, value in directories.items():
        for data in value:
            if number_document == data:
                print('Результат:\nДокумент хранится на полке: ', shelf)
                return print_shelf(dirs)
         
    if number_document != data: 
        print ('Результат:\nДокумент не найден в базе!')  
        return main()
    
#Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам
def print_docs_info(docs, dirs):
    print(f"Текущий список документов:\n")
    for shelf, data in directories.items():
        for doc in documents:
            if doc['number'] in data:
                print(f"№{doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {shelf}")
                
    return main()

#Пункт 4. Пользователь по команде “as” может добавить новую полку
def add_new_shelf(dirs): 
    number_shelf = input('Введите номер полки: ')
    for shelf in directories.keys():
        if shelf == number_shelf:
            print('Результат:\nТакая полка уже существует. Текущий перечень полок: ', *list(directories.keys()))
            return add_new_shelf(dirs)
    else: 
        shelf != number_shelf
        directories.setdefault(number_shelf)
        directories[number_shelf] = []
        print('Результат:\nПолка добавлена. Текущий перечень полок: ', *list(directories.keys()))
        return main()

#Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
def delete_empty_shelf(dirs):
    number_shelf = input('Введите номер полки: ')
    
    if number_shelf in directories:
        if len(directories[number_shelf]) != 0:
            print('Результат:\nНа полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', *list(directories.keys()))
        else:
            del directories[number_shelf]
            print('Результат:\nПолка удалена. Текущий перечень полок: ', *list(directories.keys()))
            
    else:
        print('Результат:\nТакой полки не существует. Текущий перечень полок: ', *list(directories.keys()))
    return main()

#Задание 2 (необязательное)
#Пункт 1. Пользователь по команде “ad” может добавить новый документ в данные
def add_new_document(docs, dirs): 
    number_document = input('Введите номер документа: ')
    type_document = input('Введите тип документа: ').lower()
    owner_document = input('Введите владельца документа: ').capitalize()
    number_shelf = input('Введите номер полки: ')
    new_document = {'type' : type_document, 'number' : number_document, 'name' : owner_document}
    
    
    if number_shelf in directories.keys():
        directories[number_shelf].append(number_document)
        documents.append(new_document)
        print(f"Результат:\nДокумент добавлен.\n")
        return print_docs_info(documents, directories)
        
                 
    else:
         #   shelf_doc != number_shelf
            print('Результат:\nТакой полки не существует. Добавьте полку командой as.\n')
            print_docs_info(documents, directories)
            

#Пункт 2. Пользователь по команде “d” может удалить документ из данных
def delete_document(docs, dirs): 
    documents2 = []
    number_document = input('Введите номер документа: ')
       
    for i in range(0,len(documents)-1): 
        if documents[i]['number'] != number_document:
            documents2.append(documents[i])
                       
    for data in directories.values():
        if number_document in data:
            data.remove(number_document)
            print(f"Результат:\nДокумент удален.\n")
            return print_docs_info(documents, directories)            
        
    else:    
        print('Результат:\nДокумент не найден в базе.\n')
        print_docs_info(documents, directories)  
        
#Пункт 3. Пользователь по команде “m” может переместить документ с полки на полку
def move_doc_to_shelf(docs, dirs):
    number_document = input('Введите номер документа: ')
    number_shelf = input('Введите номер полки: ')
    

    if number_shelf not in directories.keys():           
            print("Результат:\nТакой полки не существует. Текущий перечень полок:",*(list(directories.keys())),sep=", ") 
            
    elif number_document not in [doc for value in directories.values() for doc in value]:
         print("Результат:\nДокумент не найден в базе.")
    print_docs_info(documents, directories)
        
             

    def get_key(directories, number_document): # находим полку, где документ лежит изначально
        for shelf, value in directories.items():
                for doc in value:
                    if doc == number_document:            
                        return shelf     
                    
    if number_document in [doc for value in directories.values() for doc in value] and number_document not in directories[number_shelf] and number_shelf in directories.keys():
                directories[get_key(directories, number_document)].remove(number_document)
                directories[number_shelf].append(number_document)
                print("Результат:\nДокумент перемещен.\n")
                print_docs_info(documents, directories)
                
main()
```
